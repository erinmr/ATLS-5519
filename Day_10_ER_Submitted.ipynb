# Impractical Project #32: Bessie Gets Even

Objective: See http://www.usaco.org/index.php?page=viewproblem2&cpid=546 She asks FJ to count the number of different ways he can assign values to the variables so the entire expression evaluates to an even number.

**Pseudocode:**
1. Read each letter into a list. 
2. Set final_count = 0 as intial 
2. Iterate with for loops through every letter 
  - calculate b+e+s+s+i+e, g+o+e+s, m+o+o 
  - if the solution to any of the three words is not even checked with mod 2 !=0 continue 
  - final_count += 1
3. Print final count 



```python
#Test data list
letterB = [2,3]
letterE = [5]
letterS = [7]
letterI = [9,10]
letterG = [1]
letterO = [16]
letterM = [2,19]

final_count = 0 



for b in letterB:
  for e in letterE:
    for s in letterS:
      for i in letterI:
        for g in letterG:
          for o in letterO:
            for m in letterM:
              BESSIE = b+e+s+s+i+e 
              GOES = g+o+e+s
              MOO = m+o+o 


              if BESSIE %2 !=0 and GOES%2 !=0 and MOO%2 !=0:
                continue
              else:
                final_count +=1

print(final_count)
```
