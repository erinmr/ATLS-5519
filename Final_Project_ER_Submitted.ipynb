# Problem 1 - Pig Latin ðŸ·

**Objective**  

Write a program that takes a word or phrase as input and uses indexing and slicing to return its Pig Latin equivalent.

**Pig latin rules**
1. if word starts with a vowel add 'yay' to the end 
  - example 'am' -> 'amyay'
2. if word starts with a consonant take the first consonant to the end of the word and add -'ay' after that
  - example: 'dog' -> 'ogday'
3. if word starts with consonant cluster take the cluster to the end of the word and add 'ay' after that
  - example: 'scrum' -> 'umscray'

**Pseudocode**
1. User inputs a word or phrase and set to variable phrase
2. Break the phrase into words if needed. 
3. Create piglatin word variable string  
3. For each word, determine if the first letter of the word is a vowel or consonant. I used [this code](https://www.tutorialgateway.org/python-program-to-check-character-is-vowel-or-consonant/) as a base for the letter checker. 
  - if 1st letter is a vowel add 'yay' to end of word and print
  - if 1st letter is a consonant check the second letter 
    - if the second letter is a vowel move the 1st letter to the end of the word, add 'ay' after that and print.
    - if the second letter is a consonant check the third letter
      - if the third letter is a vowel, move the 1st two letters to the end of the word and add 'ay', print new word
      - if the third letter is a consonant, move the 1st three letters to the end of the word, add 'ay' and print new word or add to piglattin phrase 
4. Print pig latin phrase

- Give the user the option to change which pig latin ending to use either -ay or some custom ending

# Problem 2 - Implement Rail Fence Cipher with two rails (Zig Zag Cipher) ðŸ˜–  

**Objective:** 

Find a different cipher than the ones we discussed in class (rail and fence cipher). Write a program that encrypts and decrypts using that cipher. 

The final submission should let the user select to decrypt or encrypt, and provide space to input the relevant message and keys (as appropriate).

**About the Rail Fence Cipher from Wikipedia:**

In the rail fence cipher is a transposition cipher, where the plaintext is written downwards diagonally on successive "rails" of an imaginary fence, then moving up when the bottom rail is reached, down again when the top rail is reached, and so on until the whole plaintext is written out. The ciphertext is then read off in rows.

For example, to encrypt the message 'WE ARE DISCOVERED. RUN AT ONCE.' with 3 "rails", write the text as:

```
W . . . E . . . C . . . R . . . U . . . O . . . 
. E . R . D . S . O . E . E . R . N . T . N . E 
. . A . . . I . . . V . . . D . . . A . . . C .
```

 <br> 

(Note that spaces and punctuation are omitted.) 

Then read off the text horizontally to get the ciphertext:

WECRUOERDSOEERNTNEAIVDAC 

**Pseudocode:**
1. Read in string 
2. Strip it of punctuation, spaces and capitalize all letters
3. Create a table called rail with * for the number of rows = len message and number of columns = key 
4. Set the dir_down = false to start because our first move will be down. 
5. If dir_down = not dir_down then rail[row][col] = message[i] in the string 
5. If dir_down True add one to row 
6. else take one away from row 
7. Establish the result list 
8. For i in range of rails and j in range of (len(message)):
  - if rail[i][j] != * our filler then append rail[i][j] to the result 
9. Convert the result list to a string
10. Print string.

1. Work with spaces and punctuation too 
2. Add alternative kinds of patterns not just the straightforward zigzag 


````python
def rail_encrypt(message, key):
  import string
  message = message.translate(str.maketrans('', '', string.punctuation))
  message = message.upper().replace(" ", "")
  result_str = ''
  rail = [['*' for i in range(len(message))]for j in range(key)] #create the array of * 
  #print(rail)
  #print(message)

  dir_down = False
  row, col = 0,0

  for i in range(len(message)): 
    if (row==0) or (row==key-1):
      dir_down = not dir_down

    rail[row][col] = message[i]
    col +=1

    if dir_down:
      row +=1
    else:
      row -=1

  #print(rail)

  result = []

  for i in range(key):
    for j in range(len(message)):
      if rail[i][j] != '*':
        #print(rail)
        result.append(rail[i][j])

  result_str="".join(map(str,result))
  return result_str

message = input('Enter message to encrypt:')
key = int(input('Enter number of rails'))

rail_encrypt(message, key)
````
